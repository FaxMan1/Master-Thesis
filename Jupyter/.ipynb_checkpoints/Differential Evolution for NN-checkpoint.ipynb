```python
# Create parabola training data
xtrain = np.random.uniform(low=-1, high=np.pi, size=(30,1))  #  Between -1 and 1
ytrain = np.sin(xtrain) # xtrain ** 2 # np.sin(xtrain)

#ytrain = ytrain + np.random.normal(0.0, 0.05, ytrain.shape) # add gaussian noise


# Create test data (NEW DATA)
xtest = np.linspace(-np.pi, np.pi, 30) #  Between -1 and 1
xtest = np.array(xtest, ndmin=2).T
ytest = np.sin(xtest) #xtest ** 2 # np.sin(xtest)
#ytest = ytest + np.random.normal(0.0, 0.05, ytest.shape) # add gaussian noise
```

```python
def MSE(theta):
    
    y, yhat = theta
    
    return np.mean((y - yhat)**2)


def paraboloid(theta, param=[0,0]):
    
    x, y = theta
    h, k = param
    
    return (x - h)**2 + (y - k)**2

def ackley(theta, param=None):
    x, y = theta
    return -20.0 * np.exp(-0.2 * np.sqrt(0.5 * (x**2 + y**2))) \
           - np.exp(0.5 * (np.cos(2 * np.pi * x) + np.cos(2 * np.pi * y))) + np.e + 20
```

# Differential Evolution: Rand/1/Bin

```python
class DE():
    
    def __init__(self, objective_function, sizes, X, y, pop_size=50, F=0.5, cr=0.5):
        self.obj = objective_function
        self.sizes = sizes
        self.X = X
        self.y = y
        self.N = pop_size
        self.F = F
        self.cr = cr
        
        self.shapes_w = list(zip(sizes[:-1], sizes[1:]))
        self.shapes_b = list(zip(np.ones(len(sizes[1:]), dtype='int'), sizes[1:]))
        self.pop = self.create_population()
        self.w_dims = len(self.pop[0][0])
        self.b_dims = len(self.pop[0][1])

        pass
    
    def create_population(self):

        agents = []
        for i in range(self.N):

            w = [np.random.randn(x,y)/np.sqrt(x) for x,y in self.shapes_w]
            w = self.to_vector(w)
            b = [np.random.randn(1, y) for y in self.sizes[1:]]
            b = self.to_vector(b)
            
            agents.append((w, b))

        return agents
    
    
    def to_vector(self, weights):

        ws = []
        for w in weights:
            ws.append(w.flatten())

        return np.concatenate(ws)
    
    def to_tensor(self, v, shapes):
    
        weights = []
        prev_i = 0

        for d in shapes:
            i = prev_i + d[0] * d[1]
            w = v[prev_i:i].reshape(d)
            prev_i = i
            weights.append(w)

        return weights
    
    def NN_obj(self, agent):
    
        weights, biases = agent
        weights = self.to_tensor(weights, self.shapes_w)
        biases = self.to_tensor(biases, self.shapes_b)

        NN = NeuralNetwork(self.sizes, startweights=weights, startbiases=biases)
        yhat = NN.feedforward(self.X)

        return self.obj([self.y, yhat])
    
    
    def mutation(self, t):
        
        w = [tup[0] for tup in t]
        b = [tup[1] for tup in t]
    
        return (w[0] + self.F * (w[1] - w[2]), b[0] + self.F * (b[1] - b[2]))
    
    # binomial crossover scheme
    def crossover(self, donor, target):
        
        w_donor, b_donor = donor
        w_target, b_target = target
        
        r_w = np.random.rand(len(w_donor))
        r_b = np.random.rand(len(b_donor))
        
        trial_w = np.array([w_donor[i] if r_w[i] < self.cr else w_target[i] for i in range(len(w_donor))])
        trial_b = np.array([b_donor[i] if r_b[i] < self.cr else b_target[i] for i in range(len(b_donor))])
        return (trial_w, trial_b)
    
        
    def evolution(self, num_epochs, verbose=False):
        # evaluate the initialized population with the objective function
        obj_all = [self.NN_obj(agent) for agent in self.pop]

        # find the best agent within the initial population
        self.best_agent = self.pop[np.argmin(obj_all)]
        #self.initial_worst_agent = self.pop[np.argmax(obj_all)].copy()
        best_obj = min(obj_all)
        prev_obj = best_obj
        self.best_objs = np.zeros(num_epochs+1)
        self.best_objs[0] = best_obj
        
        for i in range(num_epochs):
            for j in range(self.N):
                
                # Random sampling from the set of all agents exluding the current one, j
                x = self.pop[j]
                choice = np.random.choice(np.delete(np.arange(self.N), j), 3, replace=False)
                a, b, c = itemgetter(*choice)(self.pop)
                
                # Mutation
                v = self.mutation([a,b,c])
                
                # Crossover
                u = self.crossover(v, x)
                
                # Selection
                obj_u = self.NN_obj(u)
                if obj_u < self.NN_obj(x):
                    self.pop[j] = u
                    obj_all[j] = obj_u
                    

            # update the current best objective function value
            best_obj = min(obj_all)
            self.best_objs[i+1] = best_obj 
            

            if best_obj < prev_obj:
                # update best agent
                self.best_agent = self.pop[np.argmin(obj_all)]
                # update previous solution to use for next iteration
                prev_obj = best_obj
            
            if verbose and i%200==0:
                # report progress at each iteration
                print('%d: cost= %.5f' % (i, best_obj))
            
        return self.best_agent
    
    

    def evaluate(self, plot_function=None, bounds=None):
        
        plt.figure(figsize=(13,8))
        plt.plot(range(len(self.best_objs)), self.best_objs)
        plt.title('Training Graph', fontsize=24)
        plt.xlabel('Iterations', fontsize=20)
        plt.ylabel('Fitness Value', fontsize=20)
        plt.show()
        
        #print(f"Worst initialization was {self.initial_worst_agent} with a cost of \
        #{np.round(self.obj(self.initial_worst_agent), 2)}.")
        
        print(f"Best agent is {self.best_agent} with a cost of \
              {np.round(self.NN_obj(self.best_agent), 30)}.")
        
        #if plot_function is not None:
           # plot_function(self.obj, self.initial_worst_agent, bounds, 'Worst Initialization')
           # plot_function(self.obj, self.best_agent, bounds, 'Best Solution')
        
        pass

```

```python
#D = DE(objective_function=MSE, sizes=[1,2048,1024,1024,1], pop_size=50, F=0.55, cr=0.85, X=xtrain, y=ytrain)
D = DE(objective_function=MSE, sizes=[1,3,2,1], pop_size=50, F=0.55, cr=0.85, X=xtrain, y=ytrain)
```

```python
D.evolution(num_epochs=4001, verbose=True)
```

# Testing of the trained network

```python
best_weights = D.to_tensor(D.best_agent[0], D.shapes_w)
best_biases = D.to_tensor(D.best_agent[1], D.shapes_b)
NN = NeuralNetwork(D.sizes, startweights=best_weights, startbiases=best_biases)

sorted_xtrain = np.sort(xtrain, axis=0)
yhat = NN.feedforward(xtest)

plt.figure(figsize=(13, 8))
#plt.plot(xtrain, ytrain, 'kx', label = 'Training Data')
plt.plot(xtest, ytest, 'bx', label = 'Test Data')
plt.plot(xtest, yhat, '-r', label = 'Learned Function')
#plt.title(r'Fit of $y = x^2$', fontsize = 15)
plt.xlabel('x', fontsize = 24)
plt.ylabel(r'$x^2$', fontsize = 24)
plt.legend(loc = 'upper center', fontsize = 14)
plt.show()
```
